```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfinv


def generate_standard_normal_samples(num_samples):
    uniform_samples = np.random.uniform(0, 1, num_samples)
    normal_samples = np.sqrt(2) * erfinv(2 * uniform_samples - 1)    
    return normal_samples

samples = generate_standard_normal_samples(1000)


plt.hist(samples, bins=30, density=True)
plt.title('Histogram of 1000 Samples Generated from Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
