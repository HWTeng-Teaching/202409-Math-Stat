```python
import numpy as np
import matplotlib.pyplot as plt

def generate_standard_normal_samples_polar(num_samples):
    samples = []
    while len(samples) < num_samples:
        u1, u2 = np.random.uniform(-1, 1, 2)
        s = u1**2 + u2**2
        if s >= 1 or s == 0:
            continue
        multiplier = np.sqrt(-2 * np.log(s) / s)
        z1 = u1 * multiplier
        z2 = u2 * multiplier
        samples.append(z1)
        if len(samples) < num_samples:
            samples.append(z2)
    
    return np.array(samples)

samples = generate_standard_normal_samples_polar(1000)


plt.hist(samples, bins=30, density=True)
plt.title('Histogram of 1000 Samples Generated from Standard Normal Distribution (Polar Method)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
