### Questions 

Write python codes to illustrate Propositions C and D (suppose $X\sim exp(1)$)  


#### Proposition C

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100000)
y = np.exp(-x)

plt.plot(x, y)

plt.title("PDF of X~Exp(1)")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

$Z = F(X) = 1 - e^{-X}$

```python

X = np.random.exponential(scale=1, size=100000)
plt.hist(X, bins=50, density=True, alpha=1)
plt.title("Random generate X")
plt.xlabel("X")
plt.ylabel("Density")
plt.show()

Z = 1 - np.exp(-X)
plt.hist(Z, bins=50, density=True, range=[-2, 3])
plt.title("Z = 1 - e^(-X)")
plt.xlabel("Z")
plt.ylabel("Density")
plt.show()
```

#### Proposition D

```python
from scipy.stats import uniform


U = np.random.uniform(0, 1, 100000)
X = uniform.ppf(U) 

x_vals = np.linspace(-1, 1, 100000)
cdf_vals = uniform.cdf(x_vals) 

plt.hist(X, bins=30, density=True, cumulative=True)

plt.title('CDF of X = F^{-1}(U)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.show()
```
