Write python codes to illustrate Propositions C and D (suppose $X\sim exp(1)$)  


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 1000)
y = np.exp(-x)

plt.plot(x, y)

plt.title("PDF of X~Exp(1)")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

$Z = F(X) = 1 - e^{-X}$

```python

X = np.random.exponential(scale=1, size=1000000)
plt.hist(X, bins=50, density=True, alpha=1)
plt.title("Random X")
plt.xlabel("X")
plt.ylabel("Density")
plt.show()

Z = 1 - np.exp(-X)
plt.hist(Z, bins=50, density=True, alpha=1, range=[-2, 3])
plt.title("Z = 1 - e^(-X)")
plt.xlabel("Z")
plt.ylabel("Density")
plt.show()
```
