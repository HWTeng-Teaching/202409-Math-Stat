One of the most commonly used (but not one of the best) methods of generating pseudorandom numbers is the linear congruential method, which works
as follows. Let x0 be an initial number (the “seed”). The sequence is generated
recursively as
$ x_n = (ax_{n−1} + c) mod \space m $


```python
#Part a
import random
repeat_time = 100
def pseudoNum_generator(a, c, m):
    count = dict()
    x0 = random.randint(1, m - 1)
    for i in range(repeat_time):
        print(x0, end=' ')
        if x0 in count:
            count[x0] += 1
        else:
            count[x0] = 1
        x0 = (a * x0 + c) % m
    print("\n")
    dict(sorted(count.items(), key=lambda item: item[1], reverse=True))
    for i in count:
        print(i,":", count[i])
a = 2
c = 3
m = 13
pseudoNum_generator(a, c, m)
```

It doesn't looks random.  
The generated number sequence is periodic with linear congruential algorithm.
